```python
import torch
import torch.nn as nn
from tqdm.auto import tqdm

from torchvision import transforms
from torch.utils.data import DataLoader

from torchvision.utils import make_grid
import matplotlib.pyplot as plt
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

## Data

```python
path = "./geant.pt" ## Write the data's path to load the data
geant = torch.load(path) 
```

```python
# Normalize data into [0,1] interval
geant = geant / geant.max()
```

```python
dataloader = DataLoader(
    geant,
    batch_size=64,
    shuffle=True
)
```

```python
def show_tensor_images(tensor_images, num_images=9, size=(1, 32, 32)):
    image_unflat = tensor_images.detach().cpu().reshape(-1, *size)
    image_grid = make_grid(image_unflat[0:num_images], nrow=3, padding=2, pad_value=.07)
    #print(image_grid[0].shape)
    #print(image_grid[0])
    plt.imshow(image_grid[0], cmap="plasma")
    plt.axis("off")
    plt.colorbar()
    plt.show()
```

## Model

### Generator

```python
class Generator(nn.Module):
    def __init__(self, z_dim=10, im_dim=1024, hidden_dim=128):
        super(Generator, self).__init__()
        self.gen = nn.Sequential(
            self.gen_block(z_dim, hidden_dim),
            self.gen_block(hidden_dim, hidden_dim*2),
            self.gen_block(hidden_dim*2, hidden_dim*4),
            self.gen_block(hidden_dim*4, hidden_dim*8),
            nn.Linear(hidden_dim*8, im_dim),
            nn.Sigmoid()
        )
        
    def gen_block(self, input_dim, output_dim):
        
        return nn.Sequential(
            nn.Linear(input_dim, output_dim),
            nn.BatchNorm1d(output_dim),
            nn.ReLU()
        )
        
    def forward(self, noise):
        
        return self.gen(noise)
```

```python
torch.manual_seed(1)
z_dim = 64 ### First hyperparameter
```

```python
generator = Generator(z_dim).to(device)
```

```python
noise = torch.randn(20, z_dim, device=device)
generator(noise).shape
```

### Noise

```python
def get_noise(n_samples, z_dim, device="cpu"):
    
    return torch.randn(n_samples, z_dim, device=device)
```

### Generator

```python
class Discriminator(nn.Module):
    def __init__(self, im_dim=1024, hidden_dim=128):
        super(Discriminator, self).__init__()
        self.disc = nn.Sequential(
            self.disc_block(im_dim, hidden_dim*4),
            self.disc_block(hidden_dim*4, hidden_dim*2),
            self.disc_block(hidden_dim*2, hidden_dim),
            nn.Linear(hidden_dim, 1)
        )
        
    def disc_block(self, input_dim, output_dim):
        
        return nn.Sequential(
            nn.Linear(input_dim, output_dim),
            nn.LeakyReLU(negative_slope=0.2) 
        )
    
    def forward(self, image):
        
        return self.disc(image)
```

```python
torch.manual_seed(2)
discriminator = Discriminator().to(device)
```

```python
image = torch.randn(5, 1024, device=device)
discriminator(image).shape
```

## Criterion

```python
criterion = nn.BCEWithLogitsLoss()
```

```python
def get_disc_loss(gen, disc, criterion, real, num_images, z_dim, device):
    noise = get_noise(num_images, z_dim, device)
    fake = gen(noise).detach()
    
    output_fake = disc(fake)
    output_real = disc(real)
    
    loss_fake = criterion(output_fake, torch.zeros_like(output_fake))
    loss_real = criterion(output_real, torch.ones_like(output_real))
    
    return (loss_fake + loss_real)/2
```

```python
get_disc_loss(generator, discriminator, criterion, torch.randn(5, 1024).to(device),
              5, z_dim, device
)
```

```python
def get_gen_loss(gen, disc, criterion, num_images, z_dim, device):
    noise = get_noise(num_images, z_dim, device)
    fake = gen(noise)
    output_fake = disc(fake)
    
    return criterion(output_fake, torch.ones_like(output_fake))
```

```python
get_gen_loss(generator, discriminator, criterion, 5, z_dim, device)
```

## Optimizer

```python
import torch.optim as optim
learning_rate = 1e-4 ### Second hyperparameters

gen_opt = optim.Adam(generator.parameters(), lr=learning_rate)
disc_opt = optim.Adam(discriminator.parameters(), lr=learning_rate)
```

## Train

```python
n_epochs = 150 ### Third hyperparameters
display_step = 1000
```

```python
mean_generator_loss = 0
mean_discriminator_loss = 0
cur_step = 0

for epoch in range(n_epochs):
    for batch in tqdm(dataloader):
        real = batch
        num_images = len(real)
        
        real = real.reshape(num_images, -1).to(device)
        
        cur_disc_loss = get_disc_loss(
            generator, discriminator, criterion, real, num_images, z_dim, device
        )
        
        disc_opt.zero_grad()
        cur_disc_loss.backward()
        disc_opt.step()
        
        cur_gen_loss = get_gen_loss(
            generator, discriminator, criterion, num_images, z_dim, device
        )
        
        gen_opt.zero_grad()
        cur_gen_loss.backward()
        gen_opt.step()
        
        mean_discriminator_loss += cur_disc_loss.item() / display_step
        mean_generator_loss += cur_gen_loss.item() / display_step
        
        if cur_step % display_step == 0 and cur_step > 0:
            print(f"Epoch: {epoch} step: {cur_step} Generator loss: {mean_generator_loss} Discriminator loss: {mean_discriminator_loss}")
                
                
            
            show_tensor_images(real)
    
            fake = generator(get_noise(num_images, z_dim, device))
            show_tensor_images(fake)
            
            mean_generator_loss = 0
            mean_discriminator_loss = 0
        cur_step += 1
```
